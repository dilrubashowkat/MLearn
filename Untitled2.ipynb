```
#all libraries to be used in image processing
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2 
```

```
#specify the folder and content image categories
dd = "/content/PetImages"
categories = ["Dog","Cat"]
```

```
#run thru the folders for displaying the images in gray scale
for items in categories:
  path = os.path.join(dd,items)
  for img in os.listdir(path):
    img_array = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
    plt.imshow(img_array,cmap='gray')
    plt.show()
    break
    break
    
    
  
```
